```python
import pandas as pd
```

```python
myfile = 'www.pangaea.de-access-anon.log.bz2'
```

```python
logData = pd.read_csv(myfile, sep='\s+',names=['hashed ID','pangea','delete1','t1','t2','log','status','size','delete2','device'],header=None)
```

```python
logData.head()
```

```python
logData['time']=logData['t1']+logData['t2']
```

```python
logDataModified = logData.drop(['delete1', 'delete2','t1','t2','pangea'], axis=1)
```

```python
logDataModified
```

```python
def string_check(log):
    if 'GET /advanced/search' in str(log): 
        if 'q=' in str(log) :
            if 'parameter' in str(log):
                return False
            else:
                return True
        else:
            return False
    else: 
        return False
```

```python
boolSearch = logDataModified['log'].apply(lambda x: string_check(x))
```

```python
searchData = logDataModified[boolSearch]
```

```python
searchData['log'][311].split('q=')[1].split(' ')[0].split('&')[0].lower()
```

```python
searchData['log'][311].lower().split('q=')[1].split(' ')[0].split('&')[0].lower()
```

```python
keyWords = searchData['log'].apply(lambda log: log.split('q=')[1].split(' ')[0].split('&')[0].lower())
```

```python
top50KeyWords = keyWords.value_counts().head(50)
```

```python
top50KeyWords.to_csv('Top_50')
```
