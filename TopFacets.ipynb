```python
import pandas as pd
```

```python
myfile = 'www.pangaea.de-access-anon.log.bz2'
```

```python
logData = pd.read_csv(myfile, sep='\s+',names=['hashed ID','pangea','delete1','t1','t2','log','status','size','delete2','device'],header=None)
```

```python
logData.head()
```

```python
logData['time']=logData['t1']+logData['t2']
```

```python
logDataModified = logData.drop(['delete1', 'delete2','t1','t2','pangea'], axis=1)
```

```python
logDataModified
```

```python
logDataModified['log'].iloc[1]
```

```python
def string_check(log):
    if 'GET /advanced/search' in str(log): 
        if 'facets=' in str(log) :
            return True
        else:
            return False
    else: 
        return False
```

```python
boolSearch = logDataModified['log'].apply(lambda x: string_check(x))
```

```python
searchData = logDataModified[boolSearch]
```

```python
searchData
```

```python
searchData['log'].iloc[1]
```

```python
facets = searchData['log'].apply(lambda log: log.split('facets=')[1].split('&')[0].split('%')[0].split(' ')[0].lower())
```

```python
topFacets = facets.value_counts()
```

```python
topFacets.to_csv('Top_Facets')
```

```python
pd.read_csv('Top_Facets',names=['Facet','Count'],header=None)
```

```python
def fstring_check(log):
    if 'GET /advanced/search' in str(log): 
        if 'f.' in str(log) :
            return True
        else:
            return False
    else: 
        return False
```

```python
fboolSearch = logDataModified['log'].apply(lambda x: fstring_check(x))
```

```python
fsearchData = logDataModified[fboolSearch]
```

```python
fsearchData
```

```python
fsearchData['log'].iloc[316].split('f.')[1].split(' ')[0].split('&')[0].split('%')[0].lower()
```

```python
f = fsearchData['log'].apply(lambda log: log.split('f.')[1].split(' ')[0].split('&')[0].lower())
```

```python
ftop = f.value_counts()
```

```python
ftop.to_csv('top_f')
```

```python
pd.read_csv('top_f',names=['F.','Count'],header=None).head(50)
```

```python
fAdvTop = fAdv.value_counts()
```
